```python
new_row = pd.DataFrame({'Name':'Prachi', 'Team':'Bihar', 'Number':4,
                        'Position':'FG', 'Age':23, 'Height':'5-3',
                        'Weight':142, 'College':'IIEST', 'Salary':200000},
                                                            index =[0])
```

```python
# Concating the new rows to the dataframe
data_csv_1 = pd.concat([new_row, data_csv]).reset_index(drop = True)
# display(data_csv)
```

```python
display(data_csv_1)
```

```python
row2 = data_csv_1.iloc[3]
print(row2)
```

```python
# Slicing rows and columns
print("Slicing only rows(till index 4):")
x1 = data_csv_1.iloc[:4, ]
print(x1, "\n")

print("Slicing rows and columns(rows=4, col 1-4, excluding 4):")
x2 = data_csv_1.iloc[:4, 1:4]
print(x2)
```

```python
start, stop, step = 0, -2, 2

# slicing till 2nd last element of a particular column
data_csv_1["Name"]= data_csv_1["Name"].str.slice(start, stop, step)
```

```python
sr = pd.Series([11, 21, 8, 18, 65, 18, 32, 10, 5, 32, None])
index_ = pd.date_range('2010-10-09 08:45', periods = 11, freq ='Y')
# set the index
sr.index = index_
result = sr.apply(lambda x : True if x>30 else False)
print(result)
```
